```python
# List all tables in samples.tpch

tpch_df = spark.sql("SHOW TABLES IN samples.tpch")
tpch_tables = [row["tableName"] for row in tpch_df.collect()]

# Build a readable query to count rows in each table
counts_query = " UNION ALL ".join(
    [
        f"""
        SELECT
            '{table}' AS table_name,
            COUNT(*) AS row_count
        FROM samples.tpch.{table}
        """
        for table in tpch_tables
    ]
)

table_counts_df = spark.sql(counts_query)

# Create a dictionary of table names and their row counts
table_counts = {row["table_name"]: row["row_count"] for row in table_counts_df.collect()}

display(table_counts_df)
```
